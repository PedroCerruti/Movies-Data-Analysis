# Data Analysis & Data Presentation - Movies Dataset

__Features/Columns__:
* **id:** The ID of the movie (clear/unique identifier).
* **title:** The Official Title of the movie.
* **tagline:** The tagline of the movie.
* **release_date:** Theatrical Release Date of the movie.
* **genres:** Genres associated with the movie.
* **belongs_to_collection:** Gives information on the movie series/franchise the particular film belongs to.
* **original_language:** The language in which the movie was originally shot in.
* **budget_musd:** The budget of the movie in million dollars.
* **revenue_musd:** The total revenue of the movie in million dollars.
* **production_companies:** Production companies involved with the making of the movie.
* **production_countries:** Countries where the movie was shot/produced in.
* **vote_count:** The number of votes by users, as counted by TMDB.
* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

## The best and the worst movies:
- Highest Revenue
- Highest Budget
- Highest Profit (=Revenue - Budget)
- Lowest Profit (=Revenue - Budget)
- Highest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10) 
- Lowest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10)
- Highest number of Votes
- Highest Rating (only movies with 10 or more Ratings)
- Lowest Rating (only movies with 10 or more Ratings)
- Highest Popularity

```python
from IPython.display import HTML
pd.options.display.max_colwidth = 200
```

```python
df_best = df[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity"]].copy()
df_best
```

```python
df_best["profit_musd"] = df.revenue_musd.sub(df.budget_musd)
df_best["return"] = df.revenue_musd.div(df.budget_musd)
```

```python
df_best
```

```python
df_best.columns = ["", "Title", "Budget", "Revenue", "Votes", 
                   "Average Rating", "Popularity", "Profit", "ROI"]
```

```python
df_best.set_index("Title", inplace = True)
```

```python
df_best
```

```python
high_rat = df_best.sort_values(by='Average Rating', ascending=False).head(5)
high_rat
```

```python
high_roi = df_best.sort_values(by='ROI', ascending=False).head(5)
high_roi
```

```python
high_rev = df_best.loc[df_best['Budget']>=5].sort_values(by='ROI', ascending=False).head(5)
high_rev
```

```python
high_rat = df_best.loc[df_best['Votes']>=10].sort_values(by='Average Rating', ascending=False).head(5)
high_rat
```

```python
df_best["Budget"].fillna(0, inplace=True)
df_best["Votes"].fillna(0, inplace=True)
df_best.info()
```

```python
#Define function to get best and worst movies
def best_worst(n, by, ascending=False, min_budget=0, min_votes=0):
    
    df2 = df_best.loc[(df_best.Budget >=min_budget) & (df_best.Votes>=min_votes),
                      ["", by]].sort_values(by=by, ascending=ascending).head(n).copy()
    
    return HTML(df2.to_html(escape=False))
```

## Most successful/popular genres

```python
df.columns
```

```python
df.genres
```

```python
gen = df.genres.str.split("|", expand=True)
gen
```

```python
gen = gen.stack().reset_index(1, drop=True).to_frame()
gen
```

```python
gen.columns = ["Genre"]
```

```python
gen = gen.merge(df[["title", "revenue_musd", "vote_average", "popularity", "release_date", "vote_count"]], 
                how="left", left_index = True, right_index=True)
gen
```

```python
agg_gen = gen.groupby("Genre").agg(Total_Revenue=("revenue_musd", "sum"),
                            Mean_Revenue = ("revenue_musd", "mean"),
                            Mean_Rating = ("vote_average", "mean"),
                            Mean_Pop = ("popularity", "mean"),
                            Total_Movies = ("Genre", "count"))
agg_gen
```

```python
agg_gen.nlargest(10, "Total_Movies")
```

```python
agg_gen.nlargest(10, "Mean_Revenue")
```

```python
agg_gen.nlargest(10, "Mean_Rating")
```

```python
agg_gen.nlargest(10, "Mean_Pop")
```

```python
agg_gen_80 = gen[gen.release_date.between("1980-01-01", "1989-12-31")].groupby("Genre").agg(Total_Revenue=("revenue_musd", "sum"),
                            Mean_Revenue = ("revenue_musd", "mean"),
                            Mean_Rating = ("vote_average", "mean"),
                            Mean_Pop = ("popularity", "mean"),
                            Total_Movies = ("Genre", "count"))
agg_gen_80
```

```python
agg_gen_90 = gen[gen.release_date.between("1990-01-01", "1999-12-31")].groupby("Genre").agg(Total_Revenue=("revenue_musd", "sum"),
                            Mean_Revenue = ("revenue_musd", "mean"),
                            Mean_Rating = ("vote_average", "mean"),
                            Mean_Pop = ("popularity", "mean"),
                            Total_Movies = ("Genre", "count"))
agg_gen_90
```

```python
fig, ax  = plt.subplots(figsize=(14,10))
agg_gen_80.Mean_Revenue.nlargest(10).plot(kind='bar', fontsize=15, rot=45, alpha=0.5, color="green")
agg_gen_90.Mean_Revenue.nlargest(10).plot(kind='bar', fontsize=15, rot=45, alpha=0.5)
plt.title("Mean Revenue (by genre and decade)", fontsize=20)
plt.legend(["1980s", "1990s"])
plt.ylabel("Revenue (in MUSD)", fontsize=15)
plt.show()
```

```python
fig, ax  = plt.subplots(figsize=(14,10))
agg_gen_80.Mean_Rating.nlargest(10).plot(kind='bar', fontsize=15, rot=45, alpha=0.5, color="green")
agg_gen_90.Mean_Rating.nlargest(10).plot(kind='bar', fontsize=15, rot=45, alpha=0.5)
plt.title("Mean Rating (by genre and decade)", fontsize=20)
plt.legend(["1980s", "1990s"])
plt.ylabel("Rating)", fontsize=15)
plt.show()
```

```python
fig, ax  = plt.subplots(figsize=(14,10))
agg_gen_80.Mean_Pop.nlargest(10).plot(kind='bar', fontsize=15, rot=45, alpha=0.5, color="green")
agg_gen_90.Mean_Pop.nlargest(10).plot(kind='bar', fontsize=15, rot=45, alpha=0.5)
plt.title("Mean Popularity (by genre and decade)", fontsize=20)
plt.legend(["1980s", "1990s"])
plt.ylabel("Popularuty)", fontsize=15)
plt.show()
```
