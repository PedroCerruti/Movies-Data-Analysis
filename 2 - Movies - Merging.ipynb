## Preparing the Data for Merge

Drop Duplicates

```python
credits.id.value_counts()
```

```python
credits.drop_duplicates(subset = 'id', inplace=True)
```

## Merging the Data

```python
df[~df.id.isin(credits.id)]
```

```python
credits[~credits.id.isin(df.id)]
```

```python
df = df.merge(credits, how='left', left_on='id', right_on='id')
```

```python
df.info()
```

## Cleaning and Transforming the new "Cast" Column: 

- Remove quotes
- Determine the cast size for all movies and add the additional column "cast_size"
- Extract all actor names from the column "cast" and overwrite "cast"
- Deal with missing values


```python
import ast
import numpy as np
```

```python
df.cast[0]
```

```python
df.cast = df.cast.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
pd.DataFrame(df.cast[0])
```

```python
df["cast_size"] = df.cast.apply(lambda x: len(x))
```

```python
df.cast_size 
```

```python
df.cast_size.value_counts(dropna=False).head(50)
```

```python
df.cast = df.cast.apply(lambda x: "|".join(i['name'] for i in x) if isinstance(x, list) else np.nan)
```

```python
df.cast
```

```python
df.cast.value_counts(dropna=False).head(20)
```

```python
df.cast.replace("", np.nan, inplace=True)
```

## Cleaning and Transforming the new "Crew" Column

- Extract the director name from the column "crew" and create the new column "director"
- Determine the crew size for all movies and add the additional column "crew_size"
- Extract all directors names from the column "cast" and add the additional column "Director"

```python
df.crew[0]
```

```python
df.crew = df.crew.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
df.crew
```

```python
pd.DataFrame(df.crew[0])
```

```python
df["crew_size"] = df.crew.apply(lambda x: len(x))
```

```python
df.crew_size 
```

```python
df.crew_size.value_counts(dropna=False).head(50)
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
df['director'] = df.crew.apply(get_director)
```

```python
df.director
```

```python
df.director.value_counts(dropna=False).head(20)
```

## Final Steps

Drop the column "crew" and __save__ the dataset in a csv-file.

```python
df.drop(columns=['crew'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.to_csv('movies_complete.csv', index=False)
```
